# Efficient summaries

What does .agg() do ?

.agg() = a shortcut to apply multiple summary functions at once.
If you want to calculate the mean and median of a column:

sales['weekly_sales'].mean()
sales['weekly_sales'].median()

You have to write two lines.
But with .agg(), you can do both in one line:

sales['weekly_sales'].agg([np.mean, np.median])

Why use .agg()?
1-To apply more than one function at the same time.
2-To use your own custom function (like IQR).



```python
import numpy as np

# Define a custom function: IQR (Interquartile Range)
def IQR(column):
    return column.quantile(0.75) - column.quantile(0.25)

# Use .agg() to apply multiple functions at once
df['amount'].agg([np.mean, np.median, IQR])
```

# Cumulative statistics

sort_values(): Sorts the data by a specific column. It’s important to sort your data before using cumsum() or cummax() if the order matters (like in time series), otherwise results might be wrong.

```python
#sort by dates
df.sort_values('bill_date')
```

cumsum(): Calculates the cumulative sum of values in a Series or DataFrame column. Each value is the sum of all previous values plus the current one.

```python
df['amount'].cumsum()
```

cummax(): Calculates the cumulative maximum. Each value is the maximum value seen so far in the Series or column.

```python
df['amount'].cummax()
```

# What is .groupby()?

The .groupby() method in pandas is used to split your data into groups based on one or more columns. After splitting, you can perform some calculations or aggregations on each group separately. 

- How does it work?

Split: The data is divided into groups based on unique values in the column(s) you specify.

Apply: You apply a function (like sum, mean, count, etc.) to each group independently.

Combine: The results from all groups are combined back into a new object (like a Series or DataFrame).



- sales_by_type = sales.groupby('type')['weekly_sales'].sum()

.groupby('type') splits data into groups by each unique store type (e.g., A, B, C).

['weekly_sales'] selects the column to aggregate.

.sum() adds up the weekly sales within each group.

# Why use .groupby()?

- It helps analyze data per group rather than overall.

- Enables easy calculation of statistics like sum, mean, count, min, max, etc., for each category or combination of categories.

- Very useful in data aggregation and summarization.

```python
df.groupby('payment_method')['amount'].sum()
```

```python
df.groupby('payment_method')['amount'].mean()
```

```python
df.groupby('payment_status')['amount'].sum()
```
