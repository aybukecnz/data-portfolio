```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.histplot(data=df, x="Released_Year", binwidth=1)
```

```python
plt.xlabel("Yayınlanma Yılı")
plt.ylabel("Film Sayısı")
plt.title("Filmlerin Yayınlanma Yılına Göre Dağılımı")
```

```python
sns.histplot(data=df, x="Released_Year")
plt.xlabel("Yayınlanma Yılı")
plt.ylabel("Film Sayısı")
plt.title("Filmlerin Yayınlanma Yılına Göre Dağılımı")
```

```python
subset= df.head(10)
sns.histplot(data=subset, x="Released_Year", binwidth=1)
plt.xlabel("Yayınlanma Yılı")
plt.ylabel("Film Sayısı")
plt.title("İlk 10 Filmin Yayınlanma Yılı Dağılımı")
```

```python
subset= df.head(100)
sns.histplot(data=subset, x="Released_Year", binwidth=1)
# X eksenindeki yazıları yana çevir
plt.xticks(rotation=90)

plt.xlabel("Yayınlanma Yılı")
plt.ylabel("Film Sayısı")
plt.title("İlk 100 Filmin Yayınlanma Yılı Dağılımı")
```
